# Assignment 2: Predicting House Prices with Scikit Learn

The first thing to do was to load the cleaned dataset using Pandas. The data is split between values which will be used to train the model, which also has the Y sale price data for each instance, and the competition data which is without the Y sale price and will be used to make predictions for the competition

```python
import numpy as np
import pandas
data = pandas.read_csv("kaggleCompetition.csv")
data = data.values
X = data[0:1460,:-1]
y = data[0:1460,-1]
X_comp = data[1460:,:-1]
y_comp = data[1460:,-1]
```

The X data is then split between training and testing subsections. The test X makes up one third of the total X data 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 0)
```

The best possible feature selection and regression min sample split was tested using GridSearchCV with a 3-fold cross validation

```python
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn import tree
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

param_grid = {'feature_selection__k': np.arange(X_train.shape[1])+1,
             'regression__min_samples_split':  list(range(2,20,1))}

clf = Pipeline([
  ('feature_selection', SelectKBest(f_regression)),
  ('regression', tree.DecisionTreeRegressor(random_state=0))
])

np.random.seed(0)
clf_grid = GridSearchCV(clf, 
                        param_grid,
                        scoring='neg_mean_squared_error',
                        cv=3 , n_jobs=1, verbose=1)
%time _ = clf_grid.fit(X_train,y_train)
```

```python
print(clf_grid.best_params_, clf_grid.best_score_)
```

```python
from sklearn import metrics
y_test_pred = clf_grid.predict(X_test)
print(metrics.mean_squared_error(y_test, y_test_pred))
```

## 3.1 GridSearch 

A model was trained using GridSearch with nested crossvalidation. The model contains an inner loop with a 3-fold crossvalidation for hyper-parameter tuning while there is an outer loop with 4-fold cross validation for model selection 

```python
from sklearn.model_selection import GridSearchCV 
from sklearn import tree
from sklearn import metrics
from sklearn.model_selection import KFold

cv_evaluation = KFold(n_splits = 4, shuffle = True , random_state = 0)
cv_evaluation

param_grid = {'max_depth' : list(range(2,40,1)),
              'min_samples_split' : list(range(2,40,1))}

np.random.seed(0)
clfGrid = GridSearchCV(tree.DecisionTreeRegressor(),
                   param_grid,
                   scoring = 'neg_mean_squared_error',
                   cv = 3,
                   n_jobs = 1, verbose = 1)

from sklearn.model_selection import cross_val_score

GridScores = -cross_val_score(clfGrid, X,y, scoring = 'neg_mean_squared_error', cv=cv_evaluation)

print(GridScores)
print(GridScores.mean(), '+-', GridScores.std())
```

The best values for the max depth and minimum sample split are computed for the GridSearch method 

```python
clfGrid.fit(X=X_train, y=y_train)
clfGrid.best_params_
```

### 3.2 Random Search

Next a model was trained using Random Search method. This method was once again done with 3-fold crossvalidation for hyper-tuning parameters and 4-fold crossvalidation for model selection. In this case an extra hyper-tuning parameter was added called criterion to determine whether it was better to use mean square error or Friedman's version of mean square error when building the decision tree  

```python
from sklearn.model_selection import RandomizedSearchCV 
from sklearn import metrics

cv_evaluation = KFold(n_splits=4, shuffle=True, random_state=0)

from scipy.stats import uniform, expon
from scipy.stats import randint as sp_randint

param_grid = {'max_depth' : sp_randint(2,40),
              'min_samples_split': sp_randint(2,40),
              'criterion' : ["mse", "friedman_mse"] }

budget = 20 

np.random.seed(0)
clfRand = RandomizedSearchCV(tree.DecisionTreeRegressor(),
                         param_grid,
                         scoring = 'neg_mean_squared_error',
                         cv = 3,
                         n_jobs = 1, verbose = 1, 
                         n_iter = budget)

from sklearn.model_selection import cross_val_score

RandScores = -cross_val_score(clfRand, X,y, scoring = 'neg_mean_squared_error', cv=cv_evaluation)

print(RandScores)
print(RandScores.mean(), '+-', RandScores.std())
```

## 3.3 Scikit Optimize

A final model was generated using Scikit Optimize using the same set up as before with 3-fold cv hyper-parameter tuning and 4-fold cv for model selection. The budget was kept the same with this model as the Random Search model to keep a level of consistency for when the models are compared. 

```python
from skopt import BayesSearchCV 
from sklearn import metrics 

from scipy.stats import randint as sp_randint

param_grid = {'max_depth' : (2,40),
              'min_samples_split' : (2,40)}

budget = 20

np.random.seed(0)
clfBayes = BayesSearchCV(tree.DecisionTreeRegressor(), 
                    param_grid, 
                    scoring = 'neg_mean_squared_error', 
                    cv = 3, 
                    n_jobs = 1, verbose = 1, 
                    n_iter = budget)

from sklearn.model_selection import cross_val_score

SciOptScores = -cross_val_score(clfBayes, X,y, scoring = 'neg_mean_squared_error', cv=cv_evaluation)

print(SciOptScores)
print(SciOptScores.mean(), '+-', SciOptScores.std())
```

# 3.4 Final Model

The models previously generated were then compared and the best model was used and was trained using all of the data. In our case the best model was the Random Search Model. The sale prices for the competition data were then predicted using this model and the final csv document containing the predicted values was produced.
